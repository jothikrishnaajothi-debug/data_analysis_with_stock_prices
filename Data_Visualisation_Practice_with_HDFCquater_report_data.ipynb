```python
import pandas as pd,numpy as np,matplotlib.pyplot as plt, seaborn as sns, plotly.graph_objects as go

df=pd.read_csv('Quote-Equity-HDFCBANK-EQ-08-12-2025-08-03-2026.csv')
print(df)

df.head()
```

```python
df.columns=df.columns.str.strip().str.replace(' ','_')
df.columns
```

```python
df['DATE']=pd.to_datetime(df['DATE'])
df=df.sort_values('DATE')
df.set_index('DATE', inplace=True)
df.head()
```

```python
df.dtypes
```

```python
columns_to_convert=['OPEN','HIGH','PREV._CLOSE','CLOSE']

for col in columns_to_convert:
    df[col]=df[col].astype(str).str.replace(',','').astype(float)

df.dtypes
```

```python
df_columns_to_convert=df[['OPEN','HIGH','LOW','CLOSE']]
pd.set_option('display.max_row',None)
print(df_columns_to_convert)
```

```python
pd.reset_option('display.max_rows')
print(df_columns_to_convert)
```

```python
plt.figure(figsize=(14,7))
plt.plot(df['CLOSE'], label='Closing Price')
plt.title('HDFC Bank - Closing Price')
plt.xlabel('DATE')
plt.ylabel('Closing Price (INR)')
plt.legend()

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,7))

plt.plot(df['CLOSE'], label='Closing Price',color='green')
plt.title('HDFC-Band-Closing-Price')
plt.xlabel('DATE')
plt.ylabel('Closing price (INR)')
plt.grid(True)
plt.legend()
plt.show()
plt.tight_layout()
```

```python
df['Daily_returns']=df['CLOSE'].pct_change()

plt.figure(figsize=(12,6))
plt.plot(df['Daily_returns'], label='Daily return')
plt.title('Daily returns of HDFC Stock')
plt.xlabel('Date')
plt.ylabel('Daily returns')
plt.legend()
plt.grid(True)
plt.show()
plt.tight_layout()
```

```python
fig = go.Figure(data=[go.Candlestick(
 x=df.index,
 open=df['OPEN'],
 high=df['HIGH'],
 low=df['LOW'],
 close=df['CLOSE']
)])
fig.update_layout(title='HDFC Bank - Candlestick Chart',
 xaxis_title='Date',
 yaxis_title='Price (INR)',
 xaxis_rangeslider_visible=True)
fig.show()
```
